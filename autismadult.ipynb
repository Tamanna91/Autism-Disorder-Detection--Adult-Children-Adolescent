```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 20)
x = data.iloc[:,0:68].values
y = data.iloc[:,69].values

x = pca.fit_transform(x)
print(x)
print(type(x))
df_x = pd.DataFrame(x, columns= ['pc1', 'pc2', 'pc3', 'pc4', 'pc5','pc6', 'pc7', 'pc8', 'pc9', 'pc10','pc11', 'pc12', 'pc13', 'pc14', 'pc15'])
print(type(df_x))
print(df_x.head(2))

explained_variance = pca.explained_variance_ratio_
print(explained_variance)
sum = np.sum(explained_variance)
print (sum)
```

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.20, random_state=0)  
```

```python
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test)  
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import classification_report 
from sklearn.metrics import confusion_matrix
#classifier= LogisticRegression(random_state=50)  
classifier= LogisticRegression() 
classifier.fit(x_train, y_train)  
y_pred= classifier.predict(x_test)
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
print(cnf_matrix) 
from sklearn import metrics
print('Accuracy: %.3f' % metrics.accuracy_score(y_test, y_pred))
# print(metrics.recall_score(y_test, y_pred))
print('Precision: %.5f' % metrics.precision_score(y_test, y_pred))
print('Recall: %.5f' % metrics.recall_score(y_test, y_pred))
print('F1 Score: %.5f' % metrics.f1_score(y_test, y_pred))
#print('F1 Score: %.5f' % metrics.specificity(y_test, y_pred))
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
specificity = tn / (tn+fp)
print('specificity: ', specificity)
#print(classification_report(y_test, y_pred, labels=[0, 1]))

from sklearn import model_selection
from sklearn.model_selection import KFold
kfold = model_selection.KFold(n_splits=10, random_state=10)
model_kfold = LogisticRegression()
results_kfold = model_selection.cross_val_score(model_kfold, x, y, cv=kfold)
print("Accuracy: %.2f%%" % (results_kfold.mean()*100.0)) 
```
