```python
data["gender"] = data["gender"].apply(lambda val : 1 if val == 'm' else 0 if val == 'f' else val)
data["jundice"] = data["jundice"].apply(lambda val : 1 if val == 'yes' else 0 if val == 'no' else val)
data["austim"] = data["austim"].apply(lambda val : 1 if val == 'yes' else 0 if val == 'no' else val)
data["used_app_before"] = data["used_app_before"].apply(lambda val : 1 if val == 'yes' else 0 if val == 'no' else val)
data['Class/ASD'] = data['Class/ASD'].apply(lambda val: 1 if val == 'YES' else 0 if val == 'NO' else 2)
data['age_desc'] = data['age_desc'].apply(lambda val : 0 if val =='4-11 years' else 1 if val == '5 and more' else val)
```

```python
'''
from sklearn.preprocessing import LabelEncoder as LE
label_encoder = LE()
data['label_relation'] = pd.Series(label_encoder.fit_transform(data['relation']))
data['label_ethnicity'] = pd.Series(label_encoder.fit_transform(data['ethnicity']))
data['label_contry_of_res'] = pd.Series(label_encoder.fit_transform(data['contry_of_res']))
'''

data['relation'] = data['relation'].apply(lambda val : 1 if val =='Parent' else 0 if val == 'others' else val)
ethnicity_cd = {'White-European':0,'Asian':1,'Middle Eastern':2,'South Asian':3,'Black':4,'Others':5}
ethnicity_cd.keys()
data['ethnicity'] = data['ethnicity'].apply(lambda val: ethnicity_cd[val] if (val) in (ethnicity_cd.keys()) else 5)
data['ethnicity'].value_counts()
country_cd = {'United Kingdom':0,'India':1,'United States':2,'Australia':3,'Jordan':4,'New Zealand':5,'Egypt':6,'Canada':7,'United Arab Emirates':8,'Others':9}
data['contry_of_res'] = data['contry_of_res'].apply(lambda val: country_cd[val] if val in country_cd.keys() else 9)
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
pca = PCA(n_components = 10)
x = data.iloc[:,1:20].values
y = data.iloc[:,21].values

x = pca.fit_transform(x)
print('REduces Feature Number:', x.shape[1])
print(x)
print(type(x))
plt.figure()
#a_std = pca.transform(transformed)
'''
plt.figure()
plt.title(label="PCA Visualization", fontsize=30, color="blue")
#plt.xlabel("Principal Component 1")
#plt.ylabel("Principal Component 2")
plt.title(label="PCA Visualization", fontsize=40, color="blue")
plt.scatter(x[:, 0], x[:, 1], c=y)

plt.figure(figsize=(8,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap='rainbow')
'''
pd.Dataframe(pca.explained_variance_ratio_).plot.bar()
plt.legend('')
plt.xlabel('Principal Components')
plt.ylabel('Explained Varience');
```

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.30, random_state=0)  
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test) 
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import classification_report 
from sklearn.metrics import confusion_matrix
classifier= LogisticRegression(random_state=20)  
#classifier= LogisticRegression() 
classifier.fit(x_train, y_train)  
y_pred= classifier.predict(x_test)
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
print(cnf_matrix) 
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score

recall = cross_val_score(classifier, x, y,cv=10, scoring='recall')
precision = cross_val_score(classifier, x, y,cv=10, scoring='precision')
accuracy = cross_val_score(classifier, x, y,cv=10, scoring='accuracy')
f1_score = cross_val_score(classifier, x, y,cv=10, scoring='f1_macro')
#print('recall:', recall)
print('Recal_final:', recall.mean()*100,'%')
#print('precision:', precision)
print('precision_final:', precision.mean()*100,'%')
#print('f1_score:', f1_score)
print('accuracy_final:', accuracy.mean()*100,'%')
print('f1_score_final:', f1_score.mean()*100,'%')
```

```python
from sklearn.ensemble import RandomForestClassifier  
classifier= RandomForestClassifier(n_estimators= 10, criterion="entropy")  
classifier.fit(x_train, y_train)  
y_pred= classifier.predict(x_test)
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix 

recall = cross_val_score(classifier, x, y,cv=10, scoring='recall')
precision = cross_val_score(classifier, x, y,cv=10, scoring='precision')
accuracy = cross_val_score(classifier, x, y,cv=10, scoring='accuracy')
f1_score = cross_val_score(classifier, x, y,cv=10, scoring='f1_macro')
#print('recall:', recall)
print('Recal_final:', recall.mean()*100,'%')
#print('precision:', precision)
print('precision_final:', precision.mean()*100,'%')
#print('f1_score:', f1_score)
print('accuracy_final:', accuracy.mean()*100,'%')
print('f1_score_final:', f1_score.mean()*100,'%')
```
